```python
columns_to_keep = ['id', 'listing_url', 'latitude', 'longitude', 'zipcode',
                   'name', 'summary', 'space', 'description', 'neighborhood_overview', 
                   'notes', 'transit', 'access', 'interaction',
                   'house_rules', 'property_type', 'room_type', 'accommodates','bathrooms', 
                   'bedrooms', 'beds', 'bed_type','amenities',
                   'nightly_price', 'price_per_stay', 'security_deposit', 'cleaning_fee',
                   'guests_included', 'extra_people', 'minimum_nights', 'maximum_nights',
                   'host_id', 'host_url', 'host_since', 'host_about', 'host_response_time', 
                   'host_response_rate', 'host_is_superhost', 'neighbourhood_cleansed',
                   'host_total_listings_count', 'host_has_profile_pic', 'host_identity_verified',
                   'number_of_reviews', 'number_of_stays', 'review_scores_rating', 
                   'review_scores_accuracy', 'review_scores_cleanliness', 
                   'review_scores_checkin','review_scores_communication', 'review_scores_location',
                   'review_scores_value', 'requires_license', 'instant_bookable',
                   'is_business_travel_ready', 'cancellation_policy',
                   'require_guest_profile_picture', 'require_guest_phone_verification'
                   ]
```

```python
sd_listings = sd_raw[columns_to_keep]
sd_listings.head(2)
```

### Cleaning Zipcode

```python
# group the missing zipcodes by neighborhood, and ffill the missing zips 
sd_listings['zipcode'] = sd_listings.groupby('neighbourhood_cleansed')['zipcode'].ffill()
sd_listings['zipcode']

# fill in last missing zipcode for the Gaslamp Quarter (92101)
sd_listings['zipcode'] = sd_listings['zipcode'].fillna('92101')
```

```python
# zipcode column contains string values ie. (92104-5446); 
print(sd_listings['zipcode'].unique())
```

```python
# drop string values after - from column 
sd_listings['zipcode'] = sd_listings['zipcode'].str.split('-').str[0]
```

```python
# convert to type int
sd_listings['zipcode'] = sd_listings['zipcode'].astype(int)
```

### Cleaning monetary columns and convert to numerical

```python
# define the money columns
money_cols = ['nightly_price', 'price_per_stay', 'security_deposit', 'cleaning_fee', 'extra_people']

# remove all $ and , 
sd_listings[money_cols] = sd_listings[money_cols].replace({'\$':'', ',':''}, regex = True)

# convert to type float
sd_listings[money_cols] = sd_listings[money_cols].astype(float)

sd_listings.tail(2)
```

### Missing values in host_response_time

```python
# unique values for host response time
display(sd_listings['host_response_time'].value_counts())

# how many are missing?
display(sd_listings['host_response_time'].isna().sum())
```

```python
# fill in missing values with N/A; treat as an ordinal 
sd_listings['host_response_time'] = sd_listings['host_response_time'].fillna('N/A')
```

### Cleaning host_response_rate

```python
# check missing
sd_listings['host_response_rate'].isna().sum()
```

```python
# fill in missing with 0s
sd_listings['host_response_rate'] = sd_listings['host_response_rate'].fillna('0%')
```

```python
# remove the % and convert to a float
sd_listings['host_response_rate'] = sd_listings['host_response_rate'].replace({'%':''}, regex = True)
sd_listings['host_response_rate'] = sd_listings['host_response_rate'].astype(float)
```

### Converting True/False Columns Into Binary & Handling Missing Values

```python
# find all columns with binary true/false values
sd_listings.columns[sd_listings.isin(['t', 'f']).any()] 
```

```python
binary_cols = sd_listings.columns[sd_listings.isin(['t', 'f']).any()].tolist()
binary_cols
```

```python
sd_listings[binary_cols].info()
```

```python
# convert all binary columns to 1, 0
sd_listings[binary_cols] = sd_listings[binary_cols].replace(to_replace=['t', 'f'], value=[1, 0])

# fill the missing values with false, 0
sd_listings[binary_cols] = sd_listings[binary_cols].fillna(0)

# convert to type int
sd_listings[binary_cols] = sd_listings[binary_cols].astype(int)
```

### Handling Missing Review Scores

```python
sd_listings['review_scores_rating'].isna().sum()
```

```python
# fill in missing review scores with the median score grouped by neighborhoods
# create a helper function to fill in missing values for each review column 
def fill_missing(df):
    """ 
    Helper function that takes in columns from a dataframe and fills in the NaNs 
    with the median grouped by neigborhood
    """
    review_cols = ['review_scores_rating', 'review_scores_accuracy',
               'review_scores_cleanliness', 'review_scores_checkin',
               'review_scores_communication', 'review_scores_location',
               'review_scores_value']
    
    for col in review_cols:
        df[col] = df[col].fillna(df.groupby('neighbourhood_cleansed')[col].transform('median'))
    
    return df 
```

```python
fill_missing(sd_listings)
```
